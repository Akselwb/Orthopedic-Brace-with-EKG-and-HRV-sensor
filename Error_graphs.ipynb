```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('rr interval data.xlsx')
bpm_error = data['error bpm']
plt.hist(bpm_error)
plt.xlabel('Percent Error')
plt.ylabel('frequency')
plt.title('Beats per Minute Percent Error')
print('bpm mean error:', round(bpm_error.mean(),4))
print('bpm std error:', round(bpm_error.std(),4))
print('bpm max error:', round(max(bpm_error),4))
```

```python
rr_error = data['rr error']
plt.hist(rr_error)
plt.xlabel('Percent Error')
plt.ylabel('frequency')
plt.title('RR Interval Percent Error')
print('rr interval mean error:', round(rr_error.mean(),4))
print('rr interval std error:', round(rr_error.std(),4))
print('rr interval max error:', round(max(rr_error),4))
```

```python
plt.bar(list(range(0,31)), data['RR interval calculated'], color = 'blue',alpha = .5, label = 'Calculated')
plt.bar(list(range(0,31)), data['RR interval from circuit'], color = 'red', alpha = .5, label = 'From Circuit')
plt.xlabel('Trial Number')
plt.ylabel('RR interval (ms)')
plt.title('RR interval Calculated vs Measured')
plt.legend(loc = 'lower left')
```

```python
plt.bar(list(range(0,31)), data['Bpm by hand'], color = 'blue', alpha = .5, label = 'By hand')
plt.bar(list(range(0,31)), data['Bpm from circuit'], color = 'red', alpha = .5, label = 'crom circuit')
plt.xlabel('Trial Number')
plt.ylabel('Beats per Minute')
plt.title('Beats per Minute Calculated vs Measured')
plt.legend(loc = 'lower left')
```
